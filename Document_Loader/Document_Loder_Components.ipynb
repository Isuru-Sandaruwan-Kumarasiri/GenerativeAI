**Loading Text Files**

```python
from langchain_community.document_loaders import TextLoader

# Initialize the TextLoader with the path to text file
loader =TextLoader("/content/Document_Loader/example_txt_file.txt")

#Load the next data
text_data =loader.load()

print(text_data)
```

**Loading Text Files From a Directory**

```python
!pip install unstructured -qU
```

```python
from langchain_community.document_loaders import DirectoryLoader

#Initialize the Directory with the path to the directory and a glob pattern for text files
loader =DirectoryLoader("/content/Document_Loader/txt_folder",glob="**/*.txt")

# Load the text data from the directoy
dataset =loader.load()

for data in dataset:
  print("-------------------------")
  print(data.page_content)
  print(data.metadata)
```

**Loading PDF Files from a Directory**

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

# Initilaize the pyPDFDirectoryLoader with the path to directory containing PDF files
loader =PyPDFDirectoryLoader("/content/Document_Loader/pdf_loader")

#Load the PDF data from the directory
dataset =loader.load()

for data in dataset :
  print("----------------------")
  print(data.page_content)
  print(data.metadata)
```
